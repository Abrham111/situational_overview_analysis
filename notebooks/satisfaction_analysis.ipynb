```python
import sys
sys.path.append('../scripts')
from eda_analysis import load_data
from user_engagement import aggregate_metrics
from experience_analytics import aggregate_per_customer
from user_satisfaction import *
```

```python
# Load the data
file_path = "../datasets/xdr_data.csv"
data = load_data(file_path)
```

```python
# Calculating the engagement score
data_eng = aggregate_metrics(data)
calculate_engagement_score(data_eng)
```

```python
# Calculating the experience score
data_exp = aggregate_per_customer(data)
calculate_experience_score(data_exp)
```

```python
# Merging the engagement and experience scores
eng_data = calculate_engagement_score(data_eng)
exp_data = calculate_experience_score(data_exp)
merged_score = pd.merge(eng_data, exp_data, on='Bearer Id', how='inner')
```

```python
# Finding the satisfaction score and top 10 satisfied customers
sats_data, top_10_sats = calculate_satisfaction(merged_score)
```

```python
# Predict the satisfaction score using regression
regression_model(sats_data)
```

```python
# K means clustering
perform_kmeans(merged_score)
```

```python
# Aggregate the average satisfaction & experience score per cluster
data = perform_kmeans(merged_score)
aggregate_scores(data)
```

```python
columns_toshow = ['Bearer Id', 'engagement_score', 'experience_score', 'satisfaction_score']
export_data = merged_score[columns_toshow]
```
